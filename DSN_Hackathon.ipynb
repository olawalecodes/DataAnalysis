```python
#dealing with missing values
def missing_values1(df_train):
    mean_round=['form_field1','form_field6', 'form_field7', 'form_field8','form_field9', 'form_field10', 'form_field11',
                'form_field12','form_field13', 'form_field14', 'form_field15','form_field16','form_field17', 'form_field18',
                'form_field19', 'form_field20','form_field26', 'form_field27', 'form_field28','form_field29', 'form_field30', 
                'form_field31', 'form_field32','form_field34', 'form_field35', 'form_field36','form_field37', 'form_field38',
                'form_field39']
    mean_cols=['form_field2', 'form_field3', 'form_field4','form_field5','form_field21', 'form_field22', 'form_field23', 
               'form_field24','form_field25','form_field33','form_field40','form_field41', 'form_field42', 'form_field43',
               'form_field44','form_field46', 'form_field48', 'form_field49',
           'form_field50']
    import math
    for col in mean_round:
        df_train[col].fillna(value=math.ceil(np.mean(df_train[col])),inplace=True)
    for col in mean_cols:
        df_train[col].fillna(value=np.mean(df_train[col]),inplace=True)
    #the column whose missing values are replaced  by the mode
    df_train['form_field45'].fillna(value=0,inplace=True)

    return df_train

df_train=missing_values1(df_train=df_train)
```

```python
#return an outlier for a missing value
def missing_values2(df_train):
    for col in df_train.columns:
        df_train[col].fillna(-999999999999,inplace=True)
    return df_train
        
#df_train=missing_values2(df_train=df_train)        
```

```python
#label_encode categorical features
from sklearn.preprocessing import LabelEncoder
lb1=LabelEncoder()
lb2=LabelEncoder()
def label_encoder(df_train):
    #df_train=pd.get_dummies(df_train.drop('default_status',axis=1))
    df_train['form_field47']=lb1.fit_transform(df_train['form_field47'])
    if 'default_status' in df_train.columns:
        df_train['default_status']=lb2.fit_transform(df_train['default_status'])
        #{0:no,1:yes}
    
    return df_train
#no particular feature has a very high correlation with the label(default_status), so some features can be removed to reduce the dimension

encoded_df=label_encoder(df_train=df_train)
#encoded_df=label_encoder(df_train=scaled_df)
```
